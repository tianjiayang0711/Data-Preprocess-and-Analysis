### 中文分词


```python
text = '中国特色社会主义是我们党领导的伟大事业，全面推进党的建设新的伟大工程，\
是这一伟大事业取得胜利的关键所在。党坚强有力，事业才能兴旺发达，国家才能繁荣稳定，人民才能\
幸福安康。党的十八大以来，我们党坚持党要管党、从严治党，凝心聚力、直击积弊、扶正\
祛邪，党的建设开创新局面，党风政风呈现新气象。习近平总书记围绕从严管党治党提出一\
系列新的重要思想，为全面推进党的建设新的伟大工程进一步指明了方向'


seg_list = jieba.cut(text)
print([x for x in seg_list])
```

```python
seg_list = jieba.cut(text,cut_all = True)
print([x for x in seg_list])
```

```python
seg_list = jieba.cut_for_search(text)
print('|'.join(seg_list))
```

### 关键词提取

```python
#基于TF-IDF：
import jieba.analyse
# topk =  前多少个
# withWeight = 是否返回词的权重
#allowPOS = 词性

keywords = jieba.analyse.extract_tags(text,topK=20,withWeight = True,allowPOS=())
print(keywords)
```

```python
#基于text-rank：


keywords = jieba.analyse.textrank(text,topK=20,withWeight = True)
print(keywords)
```

### 词性标注

```python
import jieba.posseg as peg
words = peg.cut(text)
print([x for x in words])
```
