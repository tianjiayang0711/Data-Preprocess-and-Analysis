```python
fr = open('xyj.txt','r',encoding='UTF-8')
```

```python
characters = []
stat = {}
```

```python
for line in fr:
    #去空格
    line  = line.strip()
    #若为空行，跳过本轮循环
    if len(line) == 0:
        continue
        
#     line = unicode(line)
    
    #遍历每一个字：
    for x in range(0,len(line)):
        #去除标点符号
        if line[x] in [' ', '\t', '\n', '。', '，', '(', ')', '（', '）', '：', '□', '？', '！', '《', '》', '、', '；', '“', '”', '……']:
            continue
        if line[x] not in characters:
            characters.append(line[x])
        if not stat.__contains__(line[x]):
            stat[line[x]] = 0
        
        stat[line[x]] +=1
        
print (len(characters))
print (len(stat))
```

对字典依据value排序

```python
stat = sorted(stat.items(),key = lambda d:d[1],reverse = True)
```

```python
print(stat)
```

```python
fw = open('results.csv','w',encoding = 'UTF-8')
for item in stat:
    fw.write(item[0] + ',' + str(item[1]) + '\n')
```
